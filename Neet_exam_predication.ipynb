```python
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# NEET DATA
# ----------------------------

subjects = ["Physics", "Chemistry", "Biology"]
questions = [45, 45, 90]
marks = [180, 180, 360]

registered = 2279743
appeared = 2205035
mbbs_seats = 108000

year_data = {
    "Year": [2024, 2025, 2026],
    "Registered": [2406000, 2276000, 2279743],
    "Appeared": [2333000, 2209000, 2205035]
}

# ----------------------------
# SUBJECT ANALYSIS
# ----------------------------

subject_df = pd.DataFrame({
    "Subject": subjects,
    "Questions": questions,
    "Marks": marks
})

subject_df["Weightage %"] = (
    subject_df["Marks"] /
    subject_df["Marks"].sum()
) * 100

print("\nNEET SUBJECT ANALYSIS")
print(subject_df)

# ----------------------------
# STUDENT ANALYSIS
# ----------------------------

attendance = (appeared / registered) * 100
competition = (mbbs_seats / appeared) * 100

print("\nSTUDENT STATISTICS")
print("Registered :", registered)
print("Appeared :", appeared)
print("Attendance :", round(attendance, 2), "%")
print("Chance of MBBS Seat :", round(competition, 2), "%")

# ----------------------------
# PIE CHART
# ----------------------------

plt.figure(figsize=(8,8))

plt.pie(
    marks,
    labels=subjects,
    autopct="%1.1f%%"
)

plt.title(
    "NEET Subject Weightage"
)

plt.show()

# ----------------------------
# BAR CHART
# ----------------------------

plt.figure(figsize=(8,5))

plt.bar(
    ["Registered", "Appeared"],
    [registered, appeared]
)

plt.title(
    "NEET Student Participation"
)

plt.ylabel("Students")

plt.show()

# ----------------------------
# YEAR TREND
# ----------------------------

year_df = pd.DataFrame(year_data)

plt.figure(figsize=(10,5))

plt.plot(
    year_df["Year"],
    year_df["Registered"],
    marker="o"
)

plt.plot(
    year_df["Year"],
    year_df["Appeared"],
    marker="o"
)

plt.title(
    "NEET Year-wise Trend"
)

plt.xlabel("Year")
plt.ylabel("Students")

plt.legend([
    "Registered",
    "Appeared"
])

plt.grid()

plt.show()

# ----------------------------
# SUMMARY TABLE
# ----------------------------

summary = pd.DataFrame({
    "Metric": [
        "Registered",
        "Appeared",
        "Attendance %",
        "MBBS Seats",
        "Seat Ratio %"
    ],
    "Value": [
        registered,
        appeared,
        round(attendance,2),
        mbbs_seats,
        round(competition,2)
    ]
})

print("\nSUMMARY")
print(summary)
```
